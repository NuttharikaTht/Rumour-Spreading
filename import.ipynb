```python
import os
import json
from pymongo import MongoClient
```

## Import Dataset to MongoDB

### Connect to MongoDB

```python
client = MongoClient('127.0.0.1', 27017)
db = client['putinmissing'] #Create DB
```

### Import Section

```python
root = './putinmissing'
for folders in os.listdir(root):
    folder = root +'/'+folders
    #print(fol)
    col = db[folders]
    for fol in os.listdir(folder):
        pth = folder+'/'+fol
        #print("***"+pth)
        #print(fol)
        if fol == "source-tweets":
            for file in os.listdir(pth):
                with open(os.path.join(pth,file)) as f:
                    print("***")
                    src_twt = json.load(f)
                    col.insert_one(src_twt)
        elif fol.startswith("retweets") and fol.endswith(".json") :
            for line in open(os.path.join(pth)):
                print("###")
                rt = json.loads(line)
                col.insert_one(rt)
            
```

### Checking Section

```python
i=0 
count=[]
for collection in db.list_collection_names():
    k=0
    col = db.get_collection(collection)
    for data in col.find():
        print(f"===== {i} ====== \n {data} \n")
        k+=1
    count.insert(i,k)
    i+=1
```

## Label the Annotation

```python
root = './putinmissing'
for folders in os.listdir(root):
    folder = root +'/'+folders
    #print(fol)
    col = db.get_collection(folders)
    for fol in os.listdir(folder):
        pth = folder+'/'+fol
        #print("***"+pth)
        #print(fol)
        if fol.startswith("annotation") and fol.endswith(".json") :
            for anno in open(os.path.join(pth)):
                print("===")
                label = json.loads(anno)
                #print(folders, label)
            for x in col.find():
                query = {"id" : x['id']}
                update = {"$set" : label}
                col.update_many(query,update)

            
```
